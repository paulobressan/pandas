## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('../dados/aluguel_residencial.csv', sep=';')
dados
```

```python
# Criando uma nova variavel para adicionar o valor bruto que é a soma do Valor, Condominio e IPTU
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
# Criando uma nova variavel para adicionar o valor do metro quadrado
dados['Valor m2'] = dados['Valor'] / dados['Area']
dados.head(10)
```

```python
# Arredondar o valor m2 para conter somente duas casas decimais
dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head(10)
```

```python
# Criando uma nova variavel para adicionar o valor bruto do metro quadrado e ja arredondando para 2 casas decimais
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(10)
```

```python
# Tipos de casas que temos
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# A função apply permite que nós aplique uma função para cada linha do DataFrame
# Lambda, se a palavra 'casa' estiver em algum item da lista de tipos casa, 
# então atribui 'Casa' se não 'Apartamento' 
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(10)
```

## Excluindo Variaveis

```python
# Criando um Data Frame auxiliar atraves de outro Data Frame para excluirmos variaveis
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux.head(10)
```

```python
# Primeira maneira de excluir uma variavel
del dados_aux['Valor Bruto']
```

```python
dados_aux.head(10)
```

```python
# Segunda maneira de excluir com a função pop
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux.head(10)
```

```python
# Vamos dropar mais de uma variavel informando que vamos dropar do eixo 1
# Axis 0 = Linha
# axis 1 = Coluna
dados.drop(['Valor Bruto','Valor Bruto m2'], axis=1, inplace=True)
```

```python
dados.head(10)
```

```python
# Exportar o DataFrame para o arquivo
dados.to_csv('../dados/aluguel_residencial.csv', sep=';', index=False)
```
