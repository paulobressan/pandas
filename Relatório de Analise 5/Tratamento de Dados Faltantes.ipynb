## Tratamento de Dados Faltantes

```python
import pandas as pd
```

```python
dados = pd.read_csv('../dados/aluguel_residencial.csv', sep=';')
dados.head(5)
```

```python
# Atribuir false a campos que estão nulos com o notnull
dados.notnull()
# Atribuir true a campos que estão nulos com o isnull
dados.isnull()
```

```python
# Listar informações sobre o DataFrame
# Com eles podemos identificar os campos que contem valores nulos
# Campos que contem valores nulos, Valor, Condominio, IPTU
dados.info()
```

```python
# Atribuir true para campos nulos e False para campos não nulos
dados['Valor'].isnull()
# Criando DataFrame atraves dos registros que contem nulo no campo Valor
dados[dados['Valor'].isnull()]
# Criando DataFrame atraves dos registros que contem nulo no campo Valor, Condominio e IPTU
dados[dados['Valor'].isnull() | dados['Condominio'].isnull() | dados['IPTU'].isnull()]
```

```python
# DataFrame antes de qualquer alteração
A = dados.shape[0]
# Remover registros que estão com o campo valor Nulo
# O inplace aplica a alteração no proprio dados sem ter que reatribuir o retorno de dropna
dados.dropna(subset=['Valor'], inplace=True)
# Tamanho do DataFrame depois de remover os registros com valor nulos
B = dados.shape[0]
# Quantidade de itens removidos
A - B
```

```python
# Listando registro que contem nulos no campo Valor (Validando o dropna)
dados[dados['Valor'].isnull()]
# Não contem nenhum registro com valor
```

## Tratamento de Dados Faltantes (continuação)

```python
# Criando DataFrame que contem somente os registro com Condominio nulo e listando o tamanho do DataFrame
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
# DataFrame antes de qualquer alteração
A = dados.shape[0]
# Selecionar somente os valores True de selecao no DataFrame dados
# o ~ nega o DataFrame booleano, o que é False vira True e vice-versa
dados = dados[~selecao]
# Tamanho do DataFrame depois de remover os registros com valor nulos
B = dados.shape[0]
# Quantidade de itens removidos
A - B
```

```python
# Criando DataFrame que contem somente os registro com Condominio nulo e listando o tamanho do DataFrame
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# A função fillna atribui um valor a todos os campos nulos do DataFrame
# O inplace aplica a alteração no proprio dados sem ter que reatribuir o retorno de dropna
# dados.fillna(0, inplace=True)
# Ou podemos atribuir um valor atraves de um dicionario
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
# Verificando se depois do tratamento ainda contem algum campo nulo
# Criando DataFrame que contem somente os registro com Condominio nulo e listando o tamanho do DataFrame
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Verificando se depois do tratamento ainda contem algum campo nulo
# Criando DataFrame que contem somente os registro com Condominio nulo e listando o tamanho do DataFrame
dados[dados['IPTU'].isnull()].shape[0]
```

```python
# Listando as informações do DataFrame e analisando se contem algum campo nulo
dados.info()
```

```python
# Exportando o DataFrame para um arquivo CSV com o separdor ";" e sem os index.
dados.to_csv('../dados/aluguel_residencial.csv', sep=';', index=False)
```
